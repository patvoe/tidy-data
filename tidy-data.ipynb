```python
import pandas as pd
```

```python
library = pd.read_csv('Data Tidying Examples - Sidewinder Library.csv')
library
```

## Questions: 

- What is the unit of observation? (row)
- What are the variables? (column)
- How would we tidy this?

row -> obervation -> Ausleihe

The five units of observation are 
- person
- book
- book-genre
- book-author
- book-borrowed-by-person

```python
# Creating Book table
book_df = (
    library[["Book", "Author_A", "Author_B", "Genres"]]
    .drop_duplicates()
    .reset_index(drop=True)
    .reset_index(names="Book_id")
)
book_df
```

```python
person_df= (
    library[["First Name", "Last Name", "Date of Birth","Library Card Expired?"]]
    .drop_duplicates()
    .reset_index(drop=True)
    .reset_index(names="person_id")
)
person_df
```

```python
library_df = (
    library
    .assign(
        genre_1 = lambda df: df["Genres"].str.split("; ").str.get(0),
        genre_2 = lambda df: df["Genres"].str.split("; ").str.get(1),
)
    .drop(columns = [ "Genres" ])

)

library_df
```

```python
book_df_v2 = (
    book_df
    .assign(
        genre_1 = lambda df: df["Genres"].str.split("; ").str.get(0),
        genre_2 = lambda df: df["Genres"].str.split("; ").str.get(1),
)
    .drop(columns = [ "Genres" ])

)

book_df_v2
```

```python
genre_df = ( 
    book_df_v2
    .melt(
        id_vars = [ "Book", "Book_id", "Author_A", "Author_B"],
        value_name = "genre"
    )
    .drop(columns = ["variable"])
    .dropna(subset=["genre"])
)

genre_df.sort_values(["Book_id"])
```

```python
author_df = ( 
    genre_df
    .melt(
        id_vars = [ "Book", "Book_id", "genre"],
        value_name = "author"
    )
    .drop(columns = ["variable"])
    .dropna(subset=["author"])
)

# Filter out rows with "-" in the author column
author_new_filtered = author_df[author_df["author"] != "-"]

# Sort by book-id
author_new_filtered_sorted = author_new_filtered.sort_values(["Book_id"]).reset_index(drop=True)

author_new_filtered_sorted
```
